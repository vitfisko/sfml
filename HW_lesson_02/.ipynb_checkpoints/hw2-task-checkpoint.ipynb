#### Задание 3. Фунция предобработки

Напишите функцию, которая бы

* Удаляло идентификатор `AGREEMENT_RK`
* Избавлялась от проблем с '.' и ',' в стобцах PERSONAL_INCOME, CREDIT, FST_PAYMENT, LOAN_AVG_DLQ_AMT, LOAN_MAX_DLQ_AMT
* Что-то делала с пропусками
* Кодировала категориальные признаки

В результате, ваш датафрейм должен содержать только числа и не содержать пропусков!

```python
def preproc_data(df_input):
    df_output = df_input.copy()
    df_output = df_output.drop(['AGREEMENT_RK','REG_ADDRESS_PROVINCE','FACT_ADDRESS_PROVINCE'], axis=1)
    ## Избавлялась от проблем с '.' и ',' в стобцах PERSONAL_INCOME, CREDIT, FST_PAYMENT, LOAN_AVG_DLQ_AMT, LOAN_MAX_DLQ_AMT
    
    df_output['PERSONAL_INCOME'] = df_output['PERSONAL_INCOME'].map(lambda x: x.replace(',', '.')).astype('float')
    df_output['CREDIT'] = df_output['CREDIT'].map(lambda x: x.replace(',', '.')).astype('float')
    df_output['FST_PAYMENT'] = df_output['FST_PAYMENT'].map(lambda x: x.replace(',', '.')).astype('float')
    df_output['LOAN_AVG_DLQ_AMT'] = df_output['LOAN_AVG_DLQ_AMT'].map(lambda x: x.replace(',', '.')).astype('float')
    df_output['LOAN_MAX_DLQ_AMT'] = df_output['LOAN_MAX_DLQ_AMT'].map(lambda x: x.replace(',', '.')).astype('float')
    
    ## Your Code Here
    #Counter(df['REGION_NM']).most_common(1)[0][0]
    #df_output.loc[df_output['GEN_INDUSTRY'].isnull(), 'GEN_INDUSTRY'] = df_output['GEN_INDUSTRY'].Counter(df['GEN_INDUSTRY']).most_common(1)[0][0]
    # GEN_INDUSTRY, GEN_TITLE, ORG_TP_STATE, ORG_TP_FCAPITAL, JOB_DIR, TP_PROVINCE, REGION_NM, WORK_TIME, PREVIOUS_CARD_NUM_UTILIZED
    
    df_output.loc[df_output['GEN_INDUSTRY'].isnull(), 'GEN_INDUSTRY'] = Counter(df_output['GEN_INDUSTRY']).most_common(1)[0][0]
    df_output.loc[df_output['GEN_TITLE'].isnull(), 'GEN_TITLE'] = Counter(df_output['GEN_TITLE']).most_common(1)[0][0]
    df_output.loc[df_output['ORG_TP_STATE'].isnull(), 'ORG_TP_STATE'] = Counter(df_output['ORG_TP_STATE']).most_common(1)[0][0]
    df_output.loc[df_output['ORG_TP_FCAPITAL'].isnull(), 'ORG_TP_FCAPITAL'] = Counter(df_output['ORG_TP_FCAPITAL']).most_common(1)[0][0]
    df_output.loc[df_output['JOB_DIR'].isnull(), 'JOB_DIR'] = Counter(df_output['JOB_DIR']).most_common(1)[0][0]
    df_output.loc[df_output['TP_PROVINCE'].isnull(), 'TP_PROVINCE'] = Counter(df_output['TP_PROVINCE']).most_common(1)[0][0]
    df_output.loc[df_output['REGION_NM'].isnull(), 'REGION_NM'] = Counter(df_output['REGION_NM']).most_common(1)[0][0]
    df_output.loc[df_output['WORK_TIME'].isnull(), 'WORK_TIME'] = df_output['WORK_TIME'].median()
    df_output.loc[df_output['PREVIOUS_CARD_NUM_UTILIZED'].isnull(), 'PREVIOUS_CARD_NUM_UTILIZED'] = 0.0
    
    df_output = pd.get_dummies(df_output, columns=['EDUCATION', 'JOB_DIR','MARITAL_STATUS','ORG_TP_STATE','ORG_TP_FCAPITAL','FAMILY_INCOME',\
                                                   'GEN_INDUSTRY','GEN_TITLE','POSTAL_ADDRESS_PROVINCE','TP_PROVINCE','REGION_NM'])
    
      
    return df_output
```

```python
df_preproc = df.pipe(preproc_data)

df_train_preproc = df_preproc.query('sample == "train"').drop(['sample'], axis=1)
df_test_preproc = df_preproc.query('sample == "test"').drop(['sample'], axis=1)
```

#### Задание 4. Отделите целевую переменную и остальные признаки

Должно получится:
* 2 матрицы: X и X_test
* 2 вектора: y и y_test

```python
x = df_train_preproc['TARGET']
X_test = df_train_preproc.drop('TARGET', axis=1)

y = df_test_preproc['TARGET']
Y_test = df_test_preproc.drop('TARGET', axis=1)
```

#### Задание 5. Обучение и оценка качества разных моделей

```python
from sklearn.cross_validation import train_test_split
# test_size=0.3, random_state=42

## Your Code Here
X_train, x_test, y_train, y_test = train_test_split(X_test, x, test_size=0.33, random_state=42)
```

```python
# Попробовать следующие "черные ящики": интерфейс одинаковый 
# Постепенно мы узнаем, что они делают а сейчас учимся понимать какой работает качественнее
#     fit, 
#     predict, 
#     predict_proba

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

## DecisionTreeClassifier
model_dtc = DecisionTreeClassifier()                    
model_dtc.fit(X_train, y_train)                    
predict_dtc = model_dtc.predict(x_test)
predict_proba_dtc = model_dtc.predict_proba(x_test)

## RandomForestClassifier
model_rfc = RandomForestClassifier()                    
model_rfc.fit(X_train, y_train)                    
predict_rfc = model_rfc.predict(x_test) 
predict_proba_rfc = model_rfc.predict_proba(x_test)

## LogisticRegression
model_lr = LogisticRegression()                    
model_lr.fit(X_train, y_train)                    
predict_lr = model_lr.predict(x_test) 
predict_proba_lr = model_lr.predict_proba(x_test)
```

```python
# Посчитать метрики стандартные
# accuracy, precision, recall

from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report

## Your Code Here
print('## DecisionTreeClassifier -----------')
print('precision', precision_score(y_test, predict_dtc))
print('recall', recall_score(y_test, predict_dtc))
print('accuracy', accuracy_score(y_test, predict_dtc))

print('')
print('## RandomForestClassifier -----------')
precision_rfc = precision_score(y_test, predict_rfc)
print('precision', precision_score(y_test, predict_rfc))
print('recall', recall_score(y_test, predict_rfc))
print('accuracy', accuracy_score(y_test, predict_rfc))

print('')
print('## LogisticRegression -----------')
print('precision', precision_score(y_test, predict_lr))
print('recall', recall_score(y_test, predict_lr))
print('accuracy', accuracy_score(y_test, predict_lr))

```

```python
# Визуалищировать эти метрики всех моделей на одном графике (чтоб визуально сравнить)
# Возможно вас удивит качество! Но задача подобрана специально ;) Такое качество тоже бывает

## Your Code Here
print('\n\r## DecisionTreeClassifier -----------\n\r')
print(classification_report(y_test, predict_dtc))
print('\n\r## RandomForestClassifier -----------\n\r')
print(classification_report(y_test, predict_rfc))
print('\n\r## LogisticRegression -----------\n\r')
print(classification_report(y_test, predict_lr))


```

```python
# Потроить roc-кривые всех можелей на одном графике
# Вывести roc_auc каждой моделе
# Подпишите оси и линии

## Your Code Here
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn import metrics
from sklearn.metrics import roc_auc_score, roc_curve

fpr, tpr, threshold = roc_curve(y_test, predict_proba_dtc[:,1])
auc = roc_auc_score(y_test, predict_proba_dtc[:,1])
plt.figure(figsize=(5, 5))
plt.plot(fpr,tpr,'k',label="dtc, auc="+str(auc))

fpr, tpr, threshold = roc_curve(y_test, predict_proba_rfc[:,1])
auc = roc_auc_score(y_test, predict_proba_rfc[:,1])
plt.plot(fpr,tpr,'r',label="rfc, auc="+str(auc))

fpr, tpr, threshold = roc_curve(y_test, predict_proba_lr[:,1])
auc = roc_auc_score(y_test, predict_proba_lr[:,1])
plt.plot(fpr,tpr,label="lr, auc="+str(auc))

plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))

plt.legend(loc=0)
```
